### 3-D viz of logistic regression

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact
import numpy as np
```

```python
def plot3D(theta0: int, theta1: int, decision_boundary=True): 
    """
    This looks at a 3D visuzliation with just the word counts of "beyond" and "repair"
    from our toy example, letting theta change
    
    Set decision_boundary=True to show where Logistic Regression would predict
    Y=0 versus Y=1 
    """
    theta = np.array([theta0, theta1])
    steps = 100
    repair_vals = np.linspace(0, 10, steps)
    beyond_vals = np.linspace(0, 10, steps)
    repair_grid, beyond_grid = np.meshgrid(repair_vals, beyond_vals)

    scores_3d = theta[0] * repair_grid + theta[1] * beyond_grid

    fig = plt.figure(figsize=(14, 14))
    ax = fig.add_subplot(111, projection='3d')
    
    if decision_boundary: 
        colors = np.where(scores_3d > 0, 'red', 'blue')  # Red for Z > 0, Blue for Z <= 0
        ax.plot_surface(repair_grid, beyond_grid, scores_3d, facecolors=colors, cmap='viridis')

    else: 
        ax.plot_surface(repair_grid, beyond_grid, scores_3d, cmap='viridis')

    # labels for axes
    ax.set_xlabel('Count of "repair"')
    ax.set_ylabel('Count of "beyond"')
    ax.set_zlabel('Score from Logistic Regression')
    plt.show();
```

```python
plot3D(-2, 1, decision_boundary=False)
```

```python
plot3D(-2, 1, decision_boundary=True)
```

```python
interact(plot3D, theta0=-2, theta1=1);
```
